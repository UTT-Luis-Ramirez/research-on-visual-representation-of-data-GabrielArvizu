Diagrama de Tallo y Hoja
Uso: Se utiliza para mostrar la distribución de un conjunto de datos y para identificar patrones. Cuándo se usa: Es útil cuando se tiene un conjunto de datos pequeño a mediano y se desea ver la distribución de los datos de manera rápida. Contexto: Se usa en análisis exploratorio de datos para visualizar la forma de la distribución. Cómo se crea:

Divide cada número en un "tallo" (todos los dígitos excepto el último) y una "hoja" (el último dígito).

Lista los tallos en una columna y coloca las hojas correspondientes al lado de cada tallo.

Diagrama de Caja (Box Plot)
Uso: Se utiliza para mostrar la distribución de datos basada en un resumen de cinco números: mínimo, primer cuartil, mediana, tercer cuartil y máximo. Cuándo se usa: Es útil para identificar valores atípicos y comparar distribuciones entre diferentes grupos. Contexto: Se usa en análisis estadístico para resumir y comparar distribuciones de datos. Cómo se crea:

Calcula los cinco números resumen.

Dibuja una caja desde el primer cuartil hasta el tercer cuartil.

Dibuja una línea en la mediana.

Extiende "bigotes" desde los cuartiles hasta los valores mínimos y máximos que no son atípicos.

Histograma
Uso: Se utiliza para mostrar la distribución de un conjunto de datos continuos. Cuándo se usa: Es útil cuando se tiene un conjunto de datos grande y se desea ver la distribución de los datos. Contexto: Se usa en análisis exploratorio de datos para visualizar la frecuencia de los datos en intervalos específicos. Cómo se crea:

Divide el rango de datos en intervalos (bins).

Cuenta el número de datos en cada intervalo.

Dibuja barras para cada intervalo con alturas proporcionales a las frecuencias.
